```python
import os
import random
import matplotlib.pyplot as plt
from PIL import Image
from config import *
```

```python
categories = os.listdir(DATASET_DIR)
print(f'[INFO] Number of classes: {len(categories)}')
```

```python
n = 0
n_samples_per_class = dict.fromkeys(categories, 0)

for category in categories:
    category_path = os.path.join(DATASET_DIR, category)
    n_samples_per_class[category] = len(os.listdir(category_path))

print(f'[INFO] Number of samples: {sum(n_samples_per_class.values())}')
```

```python
values = list(n_samples_per_class.values())

fig = plt.figure(figsize=(10,3))
plt.barh(categories, values)
for index, value in enumerate(values):
    plt.text(value, index, str(value))
plt.title("Number of flowers per class")
plt.xlabel("Quantity")
plt.ylabel("Category")
plt.show()
```

```python
index = 1
fig = plt.figure(figsize=(12,12))

for category in categories:
    category_path = os.path.join(DATASET_DIR, category)
    img_paths = random.sample(os.listdir(category_path), 4)
    img_paths = [os.path.join(category_path, path) for path in img_paths]
    
    for path in img_paths:
        img = Image.open(path)
        plt.subplot(5, 4, index)
        plt.axis('off')
        plt.imshow(img)
        index += 1
```
